```python
def find_all_indexes(input_str, search_str):
    # find all the positions of where a text is in a string
    # not used so dont worry
    l1 = []
    length = len(input_str)
    index = 0
    while index < length:
        i = input_str.find(search_str, index)
        if i == -1:
            return l1
        l1.append(i)
        index = i + 1
    return l1[0]
```
